```python
from itertools import product
import pandas as pd
import datetime
from dateutil.parser import parse
import random
```

```python
#Example_Start

def expand_grid(dictionary):
   return pd.DataFrame([row for row in product(*dictionary.values())], 
                       columns=dictionary.keys())

dictionary = {'color': ['red', 'green', 'blue'], 
              'vehicle': ['car', 'van', 'truck'], 
              'cylinders': [6, 8]}

expand_grid(dictionary)
#Example_End
```

```python
date1 = '2019-01-01'
dates = [(parse(date1) - datetime.timedelta(days=x)).strftime('%d-%m-%Y') for x in range(-6, 1)]
dates.sort()
```

```python
d = {'ATM': [random.randint(11011,99099) for p in range(0, 6)], 
     'Date': dates,
     'Cash': [random.randint(100000,500000) for p in range(0,2)]
    }
d
```

```python
df = expand_grid(d)
```
